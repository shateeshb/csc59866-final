```python
# Q2 Solution

import numpy as np 
input_data = np.array([3, 5])
weights = {'node_0':[2, 4], 'node_1':[4, -5], 'output':[2, 7]}

# Calculate node 0 value: node_0_value
node_0_value = (input_data * weights['node_0']).sum()

# Calculate node 1 value: node_1_value
node_1_value = (input_data * weights['node_1']).sum()

# Put node values into array: hidden_layer_outputs
hidden_layer_outputs = np.array([node_0_value, node_1_value])

# Calculate output: output
output = (hidden_layer_outputs * weights['output']).sum()

# Print output
print(output)

```

```python
# Q3 SOLUTION

import matplotlib.pyplot as plt 
import matplotlib.cm as cm
import numpy as np 
import random
random.seed(42)
# Import KMeans
from sklearn.cluster import KMeans

points = []
new_points = []

for _ in range(500):
    points.append([random.random()*10, random.random()*10])
    new_points.append([random.random()*10, random.random()*10])

points = np.array(points)
new_points = np.array(new_points)

# Create a KMeans instance with 3 clusters: model
model = KMeans(n_clusters = 3)

# Fit model to points
model.fit(points)

# Determine the cluster labels of new_points: labels
labels = model.predict(new_points)

# Assign the columns of new_points: xs and ys
xs = new_points[:,0]
ys = new_points[:,1]

colors = {0:'r', 1:'g', 2: 'b'}
label_color = [colors[l] for l in labels]
# Make a scatter plot of xs and ys, using labels to define the colors
plt.scatter(xs, ys, c=label_color , alpha=0.5)

# Assign the cluster centers: centroids
centroids = model.cluster_centers_

# Assign the columns of centroids: centroids_x, centroids_y
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]

# Make a scatter plot of centroids_x and centroids_y
plt.scatter(centroids_x, centroids_y, marker='D', s=40)
plt.show()
```

# Dataset Mini Question (1) 

Q1: Load the Iris data set, and using Gaussian Naive Bayes classifier and split train test (set random state to 42)  from sci kit learn to classify.  

Answer: 

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# Load the wine data base
iris_data = load_iris()
# Get the X and Target Labels out
iris_x = iris_data['data']
iris_target = iris_data['target']

# Split into test,train set
X_train, X_test, y_train, y_test = train_test_split( iris_x, iris_target, test_size=0.33, random_state=42)

# Set up an Gaussian NB
gnb = GaussianNB()
# Fit the training data
gnb.fit(X_train, y_train)
# Predict the X
y_predict=gnb.predict(X_test)
# Print report
print(classification_report(y_predict, y_test))
```
